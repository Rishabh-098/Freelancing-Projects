```python
# Import necessary libraries
from selenium import webdriver
import csv
import time
from bs4 import BeautifulSoup
from selenium.webdriver.chrome.options import Options
import pandas as pd

# Read the CSV file
df = pd.read_csv(r"C:\Users\HP\OneDrive\Documents\Copy of Rado_1(1).csv")

# Set Chrome options
options = webdriver.ChromeOptions()
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3")

# Initialize WebDriver
driver = webdriver.Chrome(chrome_options=options)

# Open Google homepage
driver.get("https://www.google.com/")

# Counter for tracking progress
counter = 0

# Define header for CSV file
header=["model_no","clasp type","features","movement","case size","case shape","case material","glass material","dial colour","strap material","strap colour","ean","precious stone","gender","water resistance (m)","warranty period","country of origin","price","product"]

# Open the CSV file in append mode and write header
with open("rado1.csv",'a+',encoding='UTF8' ,newline='') as f:    
    writer = csv.DictWriter(f, fieldnames = header)
    writer.writeheader()

# Iterate through each model number in the DataFrame
for i in df["Model Number"]:
    try:
        data = []  # Initialize a list to store data for each model
        counter += 1  # Increment counter
        print(counter)
        time.sleep(3)  # Wait for 3 seconds
        
        # Append "ethoswatches" to model number for search
        model = i+"ethoswatches"
        
        # Find search input element on Google homepage
        try:
            search = driver.find_element('xpath',"//input[@class='gLFyf']")
        except:
            try:
                search = driver.find_element('xpath',"//textarea[@title ='Search']")
            except:
                try:
                    search = driver.find_element('xpath',"//input[@type='text']")
                except:
                    search = driver.find_element('xpath',"//textarea[@type='text']")
        
        time.sleep(3)  # Wait for 3 seconds
        search.send_keys(model)  # Send model number to search input
        time.sleep(3)  # Wait for 3 seconds
        
        # Click on the search button
        try:
            driver.find_element("xpath","//input[@aria-label = 'Google Search']").click()
        except:
            try:
                driver.find_element("xpath",'//input[position()=1]').click()
            except:
                driver.find_element("xpath",'//input[data-ved="0ahUKEwigjJqPvbP9AhXMg2MGHcI3AHMQ4dUDCA0"]').click()
        
        time.sleep(3)  # Wait for 3 seconds
        
        # Click on the first search result
        try:
            driver.find_element('xpath',"//h3").click()
        except:
            try:
                driver.find_element('xpath',"//h3[@class='LC20lb MBeuO DKV0Md']").click()
            except:
                try:
                    x1 = driver.find_element_by_xpath("//h3[contains(text(),'Rado')]")
                    x1.click()
                except:
                    try:
                        x2 =driver.find_element_by_xpath("//h3[contains(@class, 'LC20lb')]").click()
                        x2.click()
                    except:
                        driver.find_element('xpath',"//h3[contains(@class, 'LC20lb') or contains(text(), 'Rado')]").click()
        
        time.sleep(3)  # Wait for 3 seconds
        
        # Extract HTML source of the page
        src = driver.page_source
        soup = BeautifulSoup(src,'html.parser')
        
        # Find relevant elements on the page
        td = soup.findAll('tr',class_ = "row m-0 mb-25")
        product =soup.find('span',class_= "color_9D font_24 fFamily_mrsEaves fStyle_italic text-transform-none fWeight_regular d-block").text.strip()
        price = soup.find('span',class_="price").text
        
        # Create dictionary to store data
        dict = {'model_no':i,'product':product,'price':price}
        
        # Extract data from table rows
        for i in range(len(td)):
            v = td[i].text.replace('\n',',').strip().split(",")
            ke = v[1].lower()
            if len(v) == 4:
                ro = v[2].lower()
            else:
                ro = v[3].lower()
            dict[ke]=ro
        
        data.append(dict)  # Append data dictionary to data list
        
        print(dict)  # Print dictionary
        
        # Open CSV file in append mode and write data rows
        with open("rado1.csv",'a+',encoding='UTF8' ,newline='') as f:
            writer = csv.DictWriter(f, fieldnames = header)
            writer.writerows(data)
        
    except:
        print(i)  # Print model number if an exception occurs
        pass
    
    driver.get("https://www.google.com/")  # Navigate back to Google homepage
```
